# 09 - Final Results Summary

Goal: aggregate top-k deletion results into final tables and plots.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(PROJECT_ROOT / 'src'))

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from speech_xai_project import config

project_config = config.load_config(PROJECT_ROOT / 'configs' / 'default.yaml')
DELETION_RESULTS_CSV = config.project_path(project_config['paths']['deletion_results_csv'])
```

```python
if DELETION_RESULTS_CSV.exists():
    deletion_results = pd.read_csv(DELETION_RESULTS_CSV)
    summary_table = deletion_results.groupby(['method', 'k'], as_index=False).agg(
        mean_masked_duration=('masked_duration', 'mean'),
        mean_drop=('drop', 'mean'),
        std_drop=('drop', 'std'),
        mean_relative_drop=('relative_drop', 'mean'),
        flip_rate=('prediction_flipped', 'mean'),
    )
else:
    print(f'{DELETION_RESULTS_CSV} does not exist yet. Run Notebook 07 after generating SpeechXAI and LeGrad explanations.')
    summary_table = pd.DataFrame(
        columns=['method', 'k', 'mean_masked_duration', 'mean_drop', 'std_drop', 'mean_relative_drop', 'flip_rate']
    )
summary_table
```

```python
if summary_table.empty:
    print('No deletion results available yet; skipping deletion curve plot.')
else:
    sns.lineplot(data=summary_table, x='k', y='mean_drop', hue='method', marker='o')
    plt.ylabel('Mean confidence drop')
    plt.xlabel('SpeechXAI top-k words')
    plt.title('Top-k deletion curve')
    plt.show()
```
