# 09 - Final Results Summary

Goal: aggregate deletion results into final tables and plots.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(PROJECT_ROOT / 'src'))

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from speech_xai_project import config

project_config = config.load_config(PROJECT_ROOT / 'configs' / 'default.yaml')
DELETION_RESULTS_CSV = config.project_path(project_config['paths']['deletion_results_csv'])
```

```python
deletion_results = pd.read_csv(DELETION_RESULTS_CSV)
summary_table = deletion_results.groupby(['method', 'budget'], as_index=False).agg(
    mean_drop=('drop', 'mean'),
    std_drop=('drop', 'std'),
    mean_relative_drop=('relative_drop', 'mean'),
    flip_rate=('prediction_flipped', 'mean'),
)
summary_table
```

```python
sns.lineplot(data=summary_table, x='budget', y='mean_drop', hue='method', marker='o')
plt.ylabel('Mean confidence drop')
plt.xlabel('Masking budget fraction')
plt.title('Deletion curve')
plt.show()
```
